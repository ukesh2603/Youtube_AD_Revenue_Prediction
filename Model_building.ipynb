```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error,r2_score

scaler=StandardScaler()

df=pd.read_csv(r"/Users/suriya/Ukesh_AIML_Projects/Content_Monetization_Modeler/Data/youtube_ad_revenue_dataset.csv")

df.drop_duplicates(inplace=True)
df['likes']=df['likes'].fillna(df['likes'].mean())
df['comments']=df['comments'].fillna(df['comments'].mean())
df['watch_time_minutes']=df['watch_time_minutes'].fillna(df['watch_time_minutes'].mean())
df['date']=pd.to_datetime(df['date'])
df['Year']=df['date'].dt.year
df['Month']=df['date'].dt.month
df['Day']=df['date'].dt.day
df['Hour']=df['date'].dt.hour
df['Minute']=df['date'].dt.minute
df['Seconds']=df['date'].dt.second
df['engagement_rate']=(df['likes']+df['comments'])/df['views']
df['likes_ratio']=df['likes']/df['views']

df=pd.get_dummies(df,columns=['category','device','country'],dtype='int')
df['views']=scaler.fit_transform(df[['views']])
df['likes']=scaler.fit_transform(df[['likes']])
df['watch_time_minutes']=scaler.fit_transform(df[['watch_time_minutes']])
df['video_length_minutes']=scaler.fit_transform(df[['video_length_minutes']])
df['subscribers']=scaler.fit_transform(df[['subscribers']])
df['engagement_rate']=scaler.fit_transform(df[['engagement_rate']])
df['likes_ratio']=scaler.fit_transform(df[['likes_ratio']])


df.drop(['video_id','date'],axis=1,inplace=True)

X=df.drop(['ad_revenue_usd'],axis=1)
y=df['ad_revenue_usd']

x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.15,random_state=42)

models=[LinearRegression(),DecisionTreeRegressor(),RandomForestRegressor(),KNeighborsRegressor()]

for model in models:
    model.fit(x_train,y_train)

    train_pred=model.predict(x_train)
    test_pred=model.predict(x_test)

    print("\n")
    print(f"{type(model).__name__}")
    print(f"Train MSE: {mean_squared_error(y_train,train_pred)}")
    print(f"Train R2_score: {r2_score(y_train,train_pred)}")

    print(f"Test MSE: {mean_squared_error(y_test,test_pred)}")
    print(f"Test R2_score: {r2_score(y_test,test_pred)}")
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score

scaler=StandardScaler()

df=pd.read_csv(r"/Users/suriya/Ukesh_AIML_Projects/Content_Monetization_Modeler/Data/youtube_ad_revenue_dataset.csv")

df.drop_duplicates(inplace=True)
df['likes']=df['likes'].fillna(df['likes'].mean())
df['comments']=df['comments'].fillna(df['comments'].mean())
df['watch_time_minutes']=df['watch_time_minutes'].fillna(df['watch_time_minutes'].mean())
df['date']=pd.to_datetime(df['date'])
df['Year']=df['date'].dt.year
df['Month']=df['date'].dt.month
df['Day']=df['date'].dt.day
df['Hour']=df['date'].dt.hour
df['Minute']=df['date'].dt.minute
df['Seconds']=df['date'].dt.second
df['engagement_rate']=(df['likes']+df['comments'])/df['views']
df['likes_ratio']=df['likes']/df['views']

df=pd.get_dummies(df,columns=['category','device','country'],dtype='int')
numeric_cols=['views','likes','comments','watch_time_minutes','video_length_minutes','subscribers','engagement_rate','likes_ratio']

df[numeric_cols]=scaler.fit_transform(df[numeric_cols])


df.drop(['video_id','date'],axis=1,inplace=True)

X=df.drop(['ad_revenue_usd'],axis=1)
y=df['ad_revenue_usd']

feature_names=X.columns.tolist()

x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.15,random_state=42)

model=LinearRegression()


model.fit(x_train,y_train)

train_pred=model.predict(x_train)
test_pred=model.predict(x_test)


print(f"{type(model).__name__}")
print("\n")
print(f"Train MSE: {mean_squared_error(y_train,train_pred)}")
print(f"Train R2_score: {r2_score(y_train,train_pred)}")
print("\n")
print(f"Test MSE: {mean_squared_error(y_test,test_pred)}")
print(f"Test R2_score: {r2_score(y_test,test_pred)}")
```
